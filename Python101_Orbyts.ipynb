## The Basics of Programming

Regardless of the language you use for programming, there are a few basic ideas that should guide the way you interact with the computer:
  
*   Programming is a way to give instructions to the computer
*   Traditional programming (aka not AI) relies on giving explicit, clear, and detailed instructions to the computer: the computer is exactly doing what you tell it to do and only doing what you tell it to do. It will not guess, it will not give any leeway
*   In typical circumstances, code will be deterministic - that is when executed twice in the same the code will do the same thing both times. This will stop being true when we start doing AI and some other circumstances that aren't relevant for now.
*   Google is your friend, there are loads of people who code in Python, so it is likely they have come across similar problems to you before, and asked online forums for help (stack overflow is the typical one, but sometimes Reddit or other platforms are used). So don't hesitate to search on Google how to do a specific algorithm, or how to solve a specific error.
*   Frequently used algorithms, or instructions that need to be ran several times, are typically packaged into functions, to avoid having to write it over and over again. A function is just a nickname you give to a piece of code (that can take variables or parameters) to make your life easier and avoid repeating yourself.
*   Frequently used functions are usually packaged into libraries. That is, people have written the code for you and put it freely available online for you to use. These libraries need to be installed and imported, to tell the computer where to find the piece of code to execute (again, the computer will not guess)
*   Finally, understand the value of legibility. Good programmers should document their libraries, write comment in their code, and attempt to make their code easy to read through. If your code has a problem, you want to make it as easy as possible to find and fix for the person that runs into it. And usually that person will be yourself, a few weeks or months from now. You may think you will remember what you meant, what you were trying to do, or what clever trick you decided to use; but in all likelihood you will not (trust me, I've been there). Ultimately, saving 5 minutes on the day by naming your function "thingy" or not describing in clear English what it is meant to do, will end up costing you hours in a few weeks when you need to debug it. It's not worth it.
